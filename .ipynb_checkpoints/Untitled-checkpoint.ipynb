```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
train_num = pickle.load(open('train_num','rb'))
train_num
test_num_0 = pickle.load(open('0_test_num','rb'))
test_num_1 = pickle.load(open('1_test_num','rb'))
test_num_2 = pickle.load(open('2_test_num','rb'))
acc_0 = pickle.load(open('0_acc','rb'))
acc_1 = pickle.load(open('1_acc','rb'))
acc_2 = pickle.load(open('2_acc','rb'))
total_acc = pickle.load(open('total_acc','rb'))

print('train_num_l:{}\n0_test_num_l{}\n1_test_num_l{}\n2_test_num_l{}\n0_acc{}\n1_acc{}\n2_acc{}\ntoal_acc{}'.format(train_num[-1],test_num_0[-1],test_num_1[-1],test_num_2[-1],acc_0[-1],acc_1[-1],acc_2[-1],total_acc[-1]))

fig = plt.figure()
pt1 = fig.add_subplot(111)

pt1.set_title('Tables')
pt1.set_xlabel('training num')
pt1.set_ylabel('Accuracy')
pt1.plot(train_num,acc_0,color = 'r',linestyle='-',linewidth=1,label = "0_accuracy")
pt1.plot(train_num,acc_1,color = 'b',linestyle='-',linewidth=1,label = "1_accuracy")
pt1.plot(train_num,acc_2,color = 'g',linestyle='-',linewidth=1,label = "2_accuracy")
pt1.plot(train_num,total_acc,color = 'b',linestyle='-',linewidth=1,label = "total_accuracy")
pt1.legend()

pt2 = pt1.twinx()
pt2.set_ylabel('testing nun')
pt2.plot(train_num,test_num_0,color = 'r',linestyle='--',linewidth=1,label = "0_test_num")
pt2.plot(train_num,test_num_1,color = 'b',linestyle='--',linewidth=1,label = "1_test_num")
pt2.plot(train_num,test_num_2,color = 'g',linestyle='--',linewidth=1,label = "2_test_num")
pt2.legend()
```
